```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head(20)
```

## Player Count

* Display the total number of players


```python
#counting the players
total_players = purchase_data["SN"].nunique()

#creating new dataframe
total_players_df = pd.DataFrame([{"total_players" : [total_players]}])

total_players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#performing some calcs and intitalizing variable 
unq = purchase_data["Item ID"].nunique()
avg_price = purchase_data["Price"].mean()
num_purch = purchase_data["Purchase ID"].nunique()
tot_rev = purchase_data["Price"].sum()

#creating a new summary data frame with the calcs above
new_summ_df = pd.DataFrame([{
            "Number of Unique Items" : [unq], 
            "Average Price" : "${:,.2f}".format(avg_price),
            "Number of Purchases" : [num_purch],
            "Total Revenue" : "${:,.2f}".format(tot_rev),}])

new_summ_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# grouping new data set 
grouped_demo = purchase_data.groupby("Gender")

# doing some calcs
gd = grouped_demo[["SN"]].nunique()
total = gd["SN"].sum()
percentage=gd["SN"]/total

# creating new columns
gd["percentage"]=percentage

# sorting the values
gd = gd.sort_values("percentage", ascending = False)

# adding some cosmetics
gd["percentage"] = gd["percentage"].astype(float).map("{:,.2%}".format)

# rename
Gender_Demographics = gd.rename(columns={"SN":"Total Count"})

Gender_Demographics
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# creating the bins and labels
bins =   [0, 9, 14, 19, 24, 29, 34, 39, 45]
labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# copying the dataframe
age_demo_df = purchase_data

# applying my bin and labels
age_demo_df['Age Ranges'] = pd.cut(purchase_data['Age'], bins, labels=labels) 

# counts by bin
age_demo_df = age_demo_df.groupby('Age Ranges')[['SN']].nunique()

# rename
age_demo_df = age_demo_df.rename(columns= {'SN':'Total Count'})

# total
total_count = age_demo_df['Total Count'].sum()

# creating new percentage column
age_demo_df["Percentage of Players"] = (age_demo_df['Total Count']/total_count) * 100

# adding some cosmetics
age_demo_df['Percentage of Players'] = age_demo_df['Percentage of Players'].astype(float).map("{:,.2f}%".format)


age_demo_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# making a copy of the 'purchase_data' Dataframe then grouping by the Age Range  
purchase_analysis_age_df = purchase_data

purchase_analysis_age_df = purchase_analysis_age_df.groupby('Age Ranges')[["Purchase ID"]].nunique().rename(columns={'Purchase ID' : 'Purchase Count'})

# going to approach this exercise by creating some new columns then formatting the data at the end 
#--.--------------------.--------------------.--------------------.--------------------.------------

# Average Purchase Price
avg_purchase_price = purchase_data.groupby('Age Ranges')[['Price']].mean()
purchase_analysis_age_df['Average Purchase Price'] = avg_purchase_price

# Total Purchase Value 
tot_purchase_value = purchase_data.groupby('Age Ranges')[['Price']].sum()
purchase_analysis_age_df['Total Purchase Value'] = tot_purchase_value

# Average Total Purchase Per Person
tot = purchase_data.groupby('Age Ranges')[['SN']].nunique()
avg_tot_purchase_person=purchase_analysis_age_df['Total Purchase Value']/tot['SN']
purchase_analysis_age_df['Average Total Purchase Per Person'] = avg_tot_purchase_person

# adding the cosmetics
purchase_analysis_age_df['Average Purchase Price'] = purchase_analysis_age_df['Average Purchase Price'].astype(float).map("${:,.2f}".format)
purchase_analysis_age_df['Total Purchase Value'] = purchase_analysis_age_df['Total Purchase Value'].astype(float).map("${:,.2f}".format)
purchase_analysis_age_df['Average Total Purchase Per Person'] = purchase_analysis_age_df['Average Total Purchase Per Person'].astype(float).map("${:,.2f}".format)


purchase_analysis_age_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# making a copy of my purchase data data frame
top_spenders_df = purchase_data

# creating new data frame grouped by the individual
top_spenders_df = top_spenders_df.groupby('SN')[['Purchase ID']].count().rename(columns={'Purchase ID' : 'Purchase Count'})

# create new columns 
#--.--------------------.--

#Average Purchase Price
tot = purchase_data.groupby('SN')[["Price"]].sum()
avg_purchase_price = tot['Price']/top_spenders_df['Purchase Count']
top_spenders_df['Average Purchase Price'] = avg_purchase_price

#Total Purchase Value'
top_spenders_df['Total Purchase Value'] = tot

# add cosmetics 
#purchase_analysis_age_df['Average Purchase Price'] = purchase_analysis_age_df['Average Purchase Price'].astype(float).map("${:,.2f}".format)
top_spenders_df['Average Purchase Price'] = top_spenders_df['Average Purchase Price'].astype(float).map("${:,.2f}".format)

#sorting the Total purchase by descending order then giving it an accounting format 
top_spenders_df = top_spenders_df.sort_values('Total Purchase Value', ascending = False)
top_spenders_df['Total Purchase Value'] = top_spenders_df['Total Purchase Value'].astype(float).map("${:,.2f}".format)


top_spenders_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# using iloc to retrieve requested column 
pop_items_df = purchase_data.iloc[:,[4,5,6]]

# grouping by the Item ID and Item Name
pop_items_df_group = pop_items_df.groupby(['Item ID','Item Name'])

# gathering my aggregates 
purchase_count = pop_items_df_group[['Item ID']].count().rename(columns = {'Item ID':'Purchase Count'})


# Creating new columns
#--.--------------------.--
# Average Item Price
average_item_price = pop_items_df_group[['Price']].mean()
purchase_count['Average Item Price'] = average_item_price

# Total Purchase Value'
tot_purchase_val = purchase_count['Purchase Count'] * average_item_price['Price']
purchase_count['Total Purchase Value'] = tot_purchase_val

# ordering by the purchase count  
purchase_count = purchase_count.sort_values('Purchase Count', ascending = False)

#making a copy for later sorting purpose
most_popular_items = purchase_count.copy()

# And as always, adding cosmetics 
most_popular_items['Total Purchase Value'] = most_popular_items['Total Purchase Value'].astype(float).map("${:,.2f}".format)
most_popular_items['Average Item Price'] = most_popular_items['Average Item Price'].astype(float).map("${:,.2f}".format)

most_popular_items
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order

```python
#doing a new sort 
purchase_count = purchase_count.sort_values('Total Purchase Value', ascending = False)

#making a copy
most_profitable_items = purchase_count

# adding the cosmetics
most_profitable_items['Total Purchase Value'] = purchase_count['Total Purchase Value'].astype(float).map("${:,.2f}".format)
most_profitable_items['Average Item Price'] = purchase_count['Average Item Price'].astype(float).map("${:,.2f}".format)

most_profitable_items
```
